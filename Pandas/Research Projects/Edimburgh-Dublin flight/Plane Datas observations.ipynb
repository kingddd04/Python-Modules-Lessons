# Research on Edimburgh-Dublin Flight

Let's try to extrapolate some datas from the csv file 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Edimburgh-Dublin-Gps-Datas.csv",sep = ";", header=0) # extracting datas in a dataframe
```

# Speed and Height Over Time

```python
# Clean and convert columns
speed = df["Speed (m/s)"].str.replace(",", ".").astype(float)
height = df["Altitude (m)"].str.replace(",", ".").astype(float)

# Convert speed to km/h
speed_kmh = speed * 3.6

# Create figure and first axis
fig, ax1 = plt.subplots()

# Plot speed on the left y-axis
ax1.set_xlabel("Index")  # or "Time (hours)" if you convert index
ax1.set_ylabel("Speed (km/h)", color="tab:blue")
ax1.plot(speed_kmh, color="tab:blue", label="Speed (km/h)")
ax1.tick_params(axis="y", labelcolor="tab:blue")
ax1.grid(True)

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.set_ylabel("Altitude (m)", color="tab:orange")
ax2.plot(height, color="tab:orange", label="Altitude (m)")
ax2.tick_params(axis="y", labelcolor="tab:orange")
ax2.grid(True)

plt.title("Speed and Altitude Over Time")
plt.show()
```
